## How To Analyze Data?

### .head()

It shows first N rows in the data(by default, N=5)

```python
data.head(5)
```

### .shape

It shows the total number of rows and number of columns in the dataframe

```python
data.shape
```

### .index

It provides the Index to the dataframe

```python
data.index
```

### .columns

It shows the name of each column

```python
data.columns
```

### .dtypes

It shows the datatype of each column

```python
data.dtypes
```

### .unique()

It shows all the unique values in a column. The syntax should be name of dataset followed by name of column in square brackets followed by .unique at last. It can be applied to a single column only and not to the whole dataframe.

```python
data['Weather'].unique()
```

### .nuniuqe()

It shows the total number of unique values in each column. This can be applied to the whole dataframe.

```python
data.nunique()
```

### .count

It shows the total no. of non-null values in each column. It can be applied on a single column as well as on the whole dataframe.

```python
data.count()
```

### .value_counts()

In a column, it shows all the unique values with their count. It can be applied on a single column only.

```python
data['Weather'].value_counts()
```

### .info()

Provides basic information about the dataframe.

```python
data.info()
```
